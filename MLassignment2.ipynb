```python
df=pd.read_csv('C:\\Users\\ydhar\\Downloads\\AW.csv')
df
```

```python
usefull_columns=['YearlyIncome','NumberCarsOwned','NumberChildrenAtHome','TotalChildren','Gender','Education','Occupation','HomeOwnerFlag']
df=df[usefull_columns]
df
```

```python
categorical_cols = ['Gender','Education', 'Occupation', 'HomeOwnerFlag'] 
df = pd.get_dummies(df, columns=categorical_cols)
df
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, KBinsDiscretizer


# (a) Handling Null values
df.fillna(df.mean(), inplace=True)  # Fill null values with the mean of each column

# (b) Normalization
scaler = MinMaxScaler()
numerical_cols = ['YearlyIncome', 'NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren'] 
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# (c) Discretization (Binning) on Continuous attributes
discretizer = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df[numerical_cols] = discretizer.fit_transform(df[numerical_cols])

df.head()
```
