## Q1: Questions on Basic NumPy Array

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
reversed_arr = np.flip(arr)

print("Original array:", arr)
print("Reversed array:", reversed_arr)
```

```python
array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])


flattened_array1 = array1.flatten()


flattened_array2 = array1.ravel()

print("Original array:\n", array1)
print("Flattened array (flatten()):\n", flattened_array1)
print("Flattened array (ravel()):\n", flattened_array2)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])

# Check if the arrays are equal
if np.array_equal(arr1, arr2):
    print("The arrays are equal.")
else:
    print("The arrays are not equal.")

# Check if the arrays are identical (same object)
if arr1 is arr2:
    print("The arrays are identical (same object).")
else:
    print("The arrays are not identical (different objects).")
```

```python
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

# Find unique values and their counts
unique_y, counts_y = np.unique(y, return_counts=True)

# Find the index of the maximum count
max_count_index_y = np.argmax(counts_y)


# Most frequent value
most_frequent_value_y = unique_y[max_count_index_y]
# Indices of the most frequent value
indices_y = np.where(y == most_frequent_value_y)[0]

print(f"Array y: {y}")
print(f"Most frequent value in y: {most_frequent_value_y}")
print(f"Indices of the most frequent value in y: {indices_y}")
```

```python

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
total_sum = np.sum(gfg)

# ii. Sum of all elements row-wise
row_sum = np.sum(gfg, axis=1)

# iii. Sum of all elements column-wise
column_sum = np.sum(gfg, axis=0)

# Output results
print("Sum of all elements:", total_sum)
print("Sum of all elements row-wise:", row_sum)
print("Sum of all elements column-wise:", column_sum)
```

```python

n_array = np.array([[55, 25, 15],
                    [30, 44, 2],
                    [11, 45, 77]])

# i. Sum of diagonal elements
diagonal_sum = np.trace(n_array)

# ii. Eigenvalues of the matrix
eigenvalues = np.linalg.eigvals(n_array)

# iii. Eigenvectors of the matrix
eigenvalues_full, eigenvectors = np.linalg.eig(n_array)

# iv. Inverse of the matrix
inverse_matrix = np.linalg.inv(n_array)

# v. Determinant of the matrix
determinant = np.linalg.det(n_array)

# Output results
print("Sum of diagonal elements:", diagonal_sum)
print("Eigenvalues of the matrix:", eigenvalues)
print("Eigenvectors of the matrix:\n", eigenvectors)
print("Inverse of the matrix:\n", inverse_matrix)
print("Determinant of the matrix:", determinant)
```

```python
p = np.array([[1, 2],
              [2, 3],
              [4, 5]])

q = np.array([[4, 5, 1],
              [6, 7, 2]])

result = np.dot(p, q)

print("Matrix Multiplication Result:")
print(result)

cov_matrix = np.cov([p.flatten(), q.flatten()])

print("Covariance Matrix:")
print(cov_matrix)



p1 = np.array([[1, 2], [2, 3], [4, 5]])
q1 = np.array([[4, 5, 1], [6, 7, 2]])

result1 = np.dot(p1, q1)

print("Matrix Multiplication Result:")
print(result1)

cov_matrix1 = np.cov([p1.flatten(), q1.T.flatten()])

print("Covariance Matrix:")
print(cov_matrix1)




```

## Q2: Based on NumPy Mathematics and Statistics

```python

array = np.array([[1, -2, 3], [-4, 5, -6]])

# (i) Find element-wise absolute value
abs_array = np.abs(array)
print("Element-wise absolute value:")
print(abs_array)
print()

# (ii) Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row
flattened_array = array.flatten()
print("Percentiles of flattened array:")
print("25th percentile:", np.percentile(flattened_array, 25))
print("50th percentile:", np.percentile(flattened_array, 50))
print("75th percentile:", np.percentile(flattened_array, 75))
print()

print("Percentiles for each column:")
for col in range(array.shape[1]):
    col_array = array[:, col]
    print(f"Column {col}:")
    print("25th percentile:", np.percentile(col_array, 25))
    print("50th percentile:", np.percentile(col_array, 50))
    print("75th percentile:", np.percentile(col_array, 75))
    print()

print("Percentiles for each row:")
for row in range(array.shape[0]):
    row_array = array[row, :]
    print(f"Row {row}:")
    print("25th percentile:", np.percentile(row_array, 25))
    print("50th percentile:", np.percentile(row_array, 50))
    print("75th percentile:", np.percentile(row_array, 75))
    print()

# (iii) Mean, Median and Standard Deviation of flattened array, of each column, and each row
print("Mean, Median and Standard Deviation of flattened array:")
print("Mean:", np.mean(flattened_array))
print("Median:", np.median(flattened_array))
print("Standard Deviation:", np.std(flattened_array))
print()

print("Mean, Median and Standard Deviation for each column:")
for col in range(array.shape[1]):
    col_array = array[:, col]
    print(f"Column {col}:")
    print("Mean:", np.mean(col_array))
    print("Median:", np.median(col_array))
    print("Standard Deviation:", np.std(col_array))
    print()

print("Mean, Median and Standard Deviation for each row:")
for row in range(array.shape[0]):
    row_array = array[row, :]
    print(f"Row {row}:")
    print("Mean:", np.mean(row_array))
    print("Median:", np.median(row_array))
    print("Standard Deviation:", np.std(row_array))
    print()

# (b) For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]). Find floor, ceiling and truncated value, rounded values
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor, Ceiling and Truncated values:")
print("Floor:", np.floor(a))
print("Ceiling:", np.ceil(a))
print("Truncated:", np.trunc(a))
print()

print("Rounded values:")
print("Round to nearest integer:", np.round(a))
print("Round to one decimal place:", np.round(a, 1))
```
